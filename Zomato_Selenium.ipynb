```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.zomato.com/chennai")
```

```python
details={'Res_Name':[],'Res_Rating':[],
        'Delivery_Timing':[],'Discount':[],'Food_Names':[],
        'Price':[],"Links":[],'Image_Link':[]}
```

```python
Image_Links=driver.find_elements(By.XPATH,"//img[@alt='Restaurant Card']")
for i in Image_Links:
    try:image_links=i.get_attribute('src')
    except:image_links=None
    details["Image_Link"].append(image_links)
    
Res_Name=driver.find_elements(By.CSS_SELECTOR,"h4[class='sc-1hp8d8a-0 sc-fGSyRc eBLoVw']")
for i in Res_Name:
    try:res_name=i.text
    except:res_name=None
    details["Res_Name"].append(res_name) 
    
Res_Rating=driver.find_elements(By.XPATH,"//div[@class='sc-kIXKos koPVmg']") 
for i in Res_Rating:
    try:res_rating=i.text
    except:res_rating=None
    details["Res_Rating"].append(res_rating)
    
Delivery_Timing=driver.find_elements(By.XPATH,"//div[@class='min-basic-info-right']")
for i in Delivery_Timing:
    try:delivery_timing=i.text
    except:delivery_timing=None
    details["Delivery_Timing"].append(delivery_timing)
    
Discount=driver.find_elements(By.XPATH,"//p[@class='sc-1hez2tp-0 sc-dWcDbm gcsyAc']") 
for i in Discount:
    try:discount=i.text
    except:discount=None 
    details["Discount"].append(discount)
    
Food_Names=driver.find_elements(By.XPATH,"//p[@class='sc-1hez2tp-0 sc-dCVVYJ dvZOAI']") 
for i in Food_Names:
    try:food_names=i.text
    except:food_names=None
    details["Food_Names"].append(food_names)    
    
Price=driver.find_elements(By.XPATH,"//p[@class='sc-1hez2tp-0 sc-dCVVYJ gwxNNS']")
for i in Price:
    try:price=i.text
    except:price=None
    details["Price"].append(price)
    
Links=driver.find_elements(By.XPATH,"//a[@class='sc-bvCTgw dIvKTC']")    
for i in Links:
    try:links=i.get_attribute('href')
    except:links=None 
    details["Links"].append(links)    
```

```python
for i in details.keys():
    print(i,len(details[i]))
```

```python
details1={'Res_Name':details['Res_Name'],'Res_Rating':details['Res_Rating'],
        'Food_Names':details['Food_Names'],
        'Price':details['Price'],"Links":details['Links'],'Image_Link':details['Image_Link']}
```

```python
df=pd.DataFrame(details1)
```
